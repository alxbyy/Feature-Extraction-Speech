```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load audio file
audio_file = "C:/Users/Alexander Bryan/OAF_get_happy.wav"
y, sr = librosa.load(audio_file)
```

```python
# Menghitung chroma feature
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

# Menghitung zero crossing rate
zcr = librosa.feature.zero_crossing_rate(y)

# Menghitung RMSE
rmse = librosa.feature.rms(y=y)

# Menghitung spectral centroid
spec_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)

# Menghitung MFCC
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
```

```python
# Visualisasi chroma feature
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chroma Feature')
plt.tight_layout()
plt.show()
```

```python
# Visualisasi zero crossing rate
plt.figure(figsize=(10, 4))
plt.plot(zcr[0], color='r')
plt.fill_between(range(len(zcr[0])), zcr[0], 0, color='r', alpha=0.3)
plt.title('Zero Crossing Rate')
plt.tight_layout()
plt.show()
```

```python
plt.close()
```

```python
# Visualisasi RMSE
plt.figure(figsize=(10, 4))
plt.plot(rmse[0], color='r')
plt.fill_between(range(len(rmse[0])), rmse[0], 0, color='r', alpha=0.3)
plt.title('Root Mean Square Energy (RMSE)')
plt.tight_layout()
plt.show()
```

```python
plt.close()
```

```python
# Visualisasi spectral centroid
plt.figure(figsize=(10, 4))
plt.plot(spec_centroid[0], color='r')
plt.fill_between(range(len(spec_centroid[0])), spec_centroid[0], 0, color='r', alpha=0.3)
plt.title('Spectral Centroid')
plt.tight_layout()
plt.show()
```

```python
plt.close()
```

```python
# Visualisasi MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('Mel Frequency Cepstral Coefficients (MFCC)')
plt.tight_layout()
plt.show()
```
